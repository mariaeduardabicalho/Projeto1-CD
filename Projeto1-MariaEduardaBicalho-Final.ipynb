## Desigualde informacional no Brasil : como o acesso á diferentes fontes de informações se da no território brasileiro de 2013 a 2015?


___
## Introdução


----

   O acessso a diferentes fontes de informação é muito importante para a construção de uma população atenada e com pensamento crítico. O Brasil, país cuja educação já se encontra precária, possui monopólios televisionais e grande desiguldade social , pecisa se preocupar com esta questão.
   
   Esse trabalho pretende analisar o acesso a diferentes fontes de informação no Brasil nas dieferentes situações sociais. Para isso, será analisado o PNAD 2013, 2014 e 2015 estudando as variáveis renda, condição de moradia, tv por assinatura, internet no domicílio, posse de telefone celular e acesso a internet por 3G/4G.
   
   Tv por assinatura é um importante aspecto a ser avaliado, uma vez  que a tv está muito presente na vida do brasileiro, segundo a Pesquisa Brasileira de Mídia 2015 (PBM), realizada pela Secom (Secretaria de Comunicação da Presidência da República) 95% dos brasileiro entrevistado assistem tv e desses 79% aassiste para se informar.
    
   É importante se analisar o acesso a internete uma vez que essa proporciona um acesso variadas informações do mundo todo. Além disso, está cada vez crescendo mais, e se tornando o meio de fonte informacional para cada vez mais pessoas.
    
   Tendo em vista que existe uma grande desigualdade social no Brasil, pretende-se avaliar se há também uma grande discrepância no acesso a informação. Visando analisar qualidade de informação por qualidade de vida, as variáveis de informação tecnológicas serão comparadas com a renda e a condição de moradia dos indivíduos. Prentende-se avaliar se houve uma melhora na variabilidade informacional de toda a população entre 2013 e 2015,e se os avanços tecnológicos influenciaram isso, para tal, fontes de informação mais antigas e mais novas serão comparadas.
   
  Para uma melhor avaliação, as diferentes regiões do Brasil serão análisadas entre si também relacionando com o perfil socioeconômico da região.
